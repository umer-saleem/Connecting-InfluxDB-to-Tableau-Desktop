```python
import pandas as pd
from influxdb import InfluxDBClient
from IPython.display import clear_output

# Creating connection to InfluxDB client
client = InfluxDBClient(host = 'localhost' , port = 8086)

# Database creation
client.create_database("db_1")

# Specifying the database for inserting data
client.switch_database('db_1')

# Path of CSV data to be inserted to measurement in "db_1" database
filepath = './BTC_ns.csv'

csvreader = pd.read_csv(filepath)
print(csvreader.shape)

for row_index,row in csvreader.iterrows():
    print(row_index,row)
    print("")
    json_body = [
        {
            # Specifying the measurement name within the database
            "measurement":"BTC_ns",
            "fields": 
            {
                # Specifying the columns present in the csv file
                "close": row[1],
                "high":row[2],
                "low":row[3],
                "open":row[4],
                "volume":row[5]
            }
            
        }
        
    ]
    
    # Writing data to the specified database
    client.write_points(json_body)
```
